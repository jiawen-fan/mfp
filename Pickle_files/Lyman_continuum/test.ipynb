```python
from scipy import integrate
import matplotlib.pyplot as plt
import os
import numpy as np
from matplotlib import pyplot as plt      
import pandas as pd
from scipy import special                 
import array
import scipy as sp
import scipy.interpolate
import re
import pickle as pickle
```

```python
a_factor = ['0.1204','0.1286','0.1435','0.1453']
```

```python
mfp = []
for i in a_factor:
    with open("mfp_" + str(i) +"_new.pkl",'rb') as f:
        mfp.append(pickle.load(f))
```

```python
a_factor_float =  [float(x) for x in a_factor]
```

```python
z = np.divide(1,a_factor_float) -1
```

```python
z= [7.3,6.8,6.0,5.9]
```

```python
with plt.style.context('science','notebook'):
    x = np.linspace(0,10,1000)
    for i,p in zip(mfp,z):
        plt.plot(x,i(x),label = str(p) +' z')
    plt.axhline(y=1, color='r', linestyle='-')
    plt.legend()
    plt.ylim(0,1.5)
    plt.xlabel("Mpc")
    plt.ylabel("tau")
    plt.savefig('tau_distance.png')
```
